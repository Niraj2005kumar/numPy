```python
import numpy as np
arr = np.arange(11,21) #create in indexed
print(arr)
```

```python
print(arr[:4])
```

```python
print(arr[::3]) #array slicing with step size of 3 matlab ki array se 3 step se jump karega 
```

```python
# Reverse Alternate Elements
print(arr[::-2]) #esme humne array ko reverse kiya aur alternate elements print karaye
```

```python
# Slice with Start, Stop, Step
import numpy as np
print(arr[1:7:2])
```

```python
# NUMPY INDEXING AND SLICING OF MATRIX
# A matrix is a 2D NumPy array consisting of rows and columns.
import numpy as np

arr = np.arange(1, 17).reshape(4, 4)
print(arr)
```

```python
slice = arr[3:,3:] #slicing of matrix
slice 
```

```python
arr[:,2]
```

```python
# BOOLEAN INDEXING  : - Boolean Indexing ka use kisi condition ke basis par array se values ko select ya filter karne ke liye hota hai.
```

```python
arr = np.arange(11,21)
arr
```

```python
bool_index = arr % 2 == 0 #ye condition hai ki array ke elements ko 2 se divide karne par remainder 0 aaye to wo even number hai
bool_index
```

```python

arr = arr[bool_index] #ye condition ke basis par array ke even numbers ko select kar raha hai
arr
```
