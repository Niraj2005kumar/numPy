```python
# Array generation function :- 
import numpy as np
arr = np.arange(0, 10,1)#0 se 10 tak ke numbers ko 2 ke gap me generate karne ke liye arange function ka use karte hai.,
print(arr)
```

```python
arr = np.zeros((3,4)) #3 rows and 4 columns ka array generate karne ke liye zeros function ka use karte hai.
print(arr)
```

```python
arr = np.ones((3,4)) #3 rows and 4 columns ka array generate karne ke liye ones function ka use karte hai.
print(arr)
```

```python
arr = np.linspace(1, 10, 5) #0 se 10 tak ke numbers ko 5 equal parts me divide karne ke liye linspace function ka use karte hai.
print(arr)  #esme space dekha jata hai or jitna bada number hoga utna space jyada hoga.5 ke place me 
```

```python
#random number generation function :- NumPy me Random Generation Functions ka use random numbers generate karne ke liye hota hai
np.random.rand(3)
```

```python
# Array Attribute :-Array Attributes ka matlab hai array ki properties jaise shape, size, dimensions, data type, etc.
arr = np.array([[1, 2, 3], [4, 5, 6]])
print(arr) #shape ka matlab hai array ke dimensions, jaise rows aur columns ki sankhya.
```

```python

print(arr.shape) #shape ka matlab hai array ke dimensions, jaise rows aur columns ki sankhya.
```

```python
print(arr.size)#esse size batata hai 
```

```python
print(arr.dtype) #esse data type batata hai
```

```python
#Array method :- NumPy Array Methods wo functions hote hain jo array par operations perform karte hain.

print(arr.max)
```

```python
arr.sum() 
```

```python
np.sum(arr,axis = 0) #axis = 0 ka matlab hai ki hum rows ke across sum kar rahe hain, yani har column ka sum nikal rahe hain.
```

```python
arr.mean() #mean ka matlab hai average, yani array ke elements ka average nikalna.
```

```python
# argmax ka matlab hai array me sabse bada element ka index nikalna.
arr = np.array([[1, 2, 3], [4, 5, 6]])  # define a fallback array (or run the cell that sets `arr`)
print(arr.argmax())
```
