```python
# Stack :- Stacking is a technique used in NumPy to combine multiple arrays into a single array along a specified axis. It allows you to create higher-dimensional arrays by stacking lower-dimensional arrays together.
```

```python
a = np.array([[1,2,3,4,5,6,]])
b = np.array([[7,8,9,10,11,12]])
```

```python
np.vstack((a,b))
```

```python
np.hstack((a,b))
```

```python
np.column_stack((a,b))
```

```python
# spliting arrays :- Splitting arrays in NumPy refers to the process of dividing an array into multiple smaller arrays along a specified axis. This can be useful for various purposes, such as data preprocessing, parallel processing, or simply organizing data into manageable chunks.
```

```python
c = np.arange(16).reshape(4,4)
c
```

```python
np.hsplit(c,2)
```
