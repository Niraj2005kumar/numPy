```python
import numpy as np
#columns : [Age ,Maths Marks, Science Marks,]
data = np.array([[18,85,78],
                [19,92,88],
                [17 ,76,78],
                [18,89,90],
                [19,95,92],
                [17,80,85],
                [18,88,87],
                [19,91,89],
                [17,75,80],
                [18,86,84]])
```

```python
#Get the Shape of the matrix
data.shape
```

```python
#Find the average of each student
np.mean(data[:,0])
```

```python
# Extrect Maths marks of all students
data[:,1]
```

```python
#Find the highest marks in Science
np.max(data[:,2])
```

```python
#Get details of student who scored more than 90 in maths

data[data[:,1] > 90]
```

```python
#increase the maths marks of all students by 5
data[:,1] + 5
data
```

```python
# Find how many student are younger than 19

len(data[data[:,0] < 19])
```
